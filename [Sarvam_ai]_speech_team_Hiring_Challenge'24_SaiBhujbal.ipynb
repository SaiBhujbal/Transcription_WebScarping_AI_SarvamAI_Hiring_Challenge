````python
# Install all the necessary dependencies

!apt-get install -y espeak libmagic-dev ffmpeg
!wget https://raw.githubusercontent.com/readbeyond/aeneas/master/install_dependencies.sh
!bash install_dependencies.sh
!pip install pytube aeneas pyAudioAnalysis hmmlearn eyed3 pydub moviepy gradio
!pip install git+https://github.com/openai/whisper.git

import os
import json
import nltk
import torch
import torchaudio
from pytube import YouTube
from aeneas.executetask import ExecuteTask
from aeneas.task import Task
from transformers import Wav2Vec2ForCTC, Wav2Vec2Processor
from pyAudioAnalysis import audioSegmentation as aS
import whisper

# Download NLTK data for sentence tokenization
nltk.download('punkt')

def download_video(youtube_url):
    """
    Download video from YouTube using pytube.

    Parameters:
    youtube_url (str): URL of the YouTube video.

    Returns:
    str: Path to the downloaded video file or error message.
    """
    try:
        # Create YouTube object
        yt = YouTube(youtube_url)
        # Get the first available video stream with both video and audio
        video_stream = yt.streams.filter(progressive=True, file_extension='mp4').first()
        # Download the video and save it as 'video.mp4'
        video_path = video_stream.download(filename='video.mp4')
        return video_path
    except Exception as e:
        # Return error message if an exception occurs
        return str(e)

def extract_audio(video_path):
    """
    Extract audio from the downloaded video using ffmpeg.

    Parameters:
    video_path (str): Path to the downloaded video file.

    Returns:
    str: Path to the extracted audio file or error message.
    """
    try:
        # Define the audio path by replacing the video file extension with .wav
        audio_path = video_path.replace('.mp4', '.wav')
        # Use ffmpeg to extract audio from the video
        os.system(f"ffmpeg -i {video_path} -q:a 0 -map a {audio_path}")
        return audio_path
    except Exception as e:
        # Return error message if an exception occurs
        return str(e)

def resample_audio(audio_path, target_sample_rate=16000):
    """
    Resample the extracted audio to the target sample rate using torchaudio.

    Parameters:
    audio_path (str): Path to the extracted audio file.
    target_sample_rate (int): Target sample rate for resampling (default: 16000).

    Returns:
    str: Path to the resampled audio file or error message.
    """
    try:
        # Load the audio file
        waveform, sample_rate = torchaudio.load(audio_path)
        # Create a resampler to resample the audio to the target sample rate
        resampler = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=target_sample_rate)
        # Apply the resampler to the waveform
        resampled_waveform = resampler(waveform)
        # Define the path for the resampled audio
        resampled_audio_path = "resampled_audio.wav"
        # Save the resampled audio
        torchaudio.save(resampled_audio_path, resampled_waveform, target_sample_rate)
        return resampled_audio_path
    except Exception as e:
        # Return error message if an exception occurs
        return str(e)

def transcribe_audio(audio_path):
    """
    Transcribe the audio using the Whisper model.

    Parameters:
    audio_path (str): Path to the resampled audio file.

    Returns:
    str: Transcription text or error message.
    """
    try:
        # Load the Whisper model
        model = whisper.load_model("tiny")
        # Transcribe the audio file
        result = model.transcribe(audio_path)
        # Extract the transcription text from the result
        transcription = result["text"]
        return transcription
    except Exception as e:
        # Return error message if an exception occurs
        return str(e)

def align_transcript(audio_path, transcript):
    """
    Align the transcript with the audio using aeneas.

    Parameters:
    audio_path (str): Path to the resampled audio file.
    transcript (str): Transcription text.

    Returns:
    str: Path to the alignment JSON file or error message.
    """
    try:
        # Save the transcript to a text file
        transcript_path = 'transcript.txt'
        with open(transcript_path, 'w') as f:
            f.write(transcript)

        # Define the aeneas task configuration
        config_string = "task_language=eng|is_text_type=plain|os_task_file_format=json"
        task = Task(config_string=config_string)
        task.audio_file_path_absolute = audio_path
        task.text_file_path_absolute = transcript_path
        task.sync_map_file_path_absolute = "alignment.json"

        # Execute the aeneas task
        ExecuteTask(task).execute()
        # Save the alignment result to a JSON file
        task.output_sync_map_file()
        return "alignment.json"
    except Exception as e:
        # Return error message if an exception occurs
        return str(e)

def semantic_chunking(alignment_path):
    """
    Perform semantic chunking of the transcript based on the alignment.

    Parameters:
    alignment_path (str): Path to the alignment JSON file.

    Returns:
    list: List of semantic chunks or error message.
    """
    try:
        # Load the alignment JSON file
        with open(alignment_path, 'r') as f:
            alignment = json.load(f)

        chunks = []
        chunk_id = 0

        # Iterate through each fragment in the alignment
        for fragment in alignment['fragments']:
            start_time = float(fragment['begin'])
            end_time = float(fragment['end'])
            text = fragment['lines'][0] if fragment['lines'] else ""
            chunk_length = end_time - start_time

            if chunk_length > 15:
                # Split long chunks into smaller chunks
                words = text.split()
                num_words = len(words)
                words_per_chunk = max(1, int(num_words / (chunk_length / 15)))
                for i in range(0, num_words, words_per_chunk):
                    chunk_text = " ".join(words[i:i + words_per_chunk])
                    chunks.append([
                        chunk_id,
                        15,
                        chunk_text,
                        start_time + (i / num_words) * chunk_length,
                        start_time + ((i + words_per_chunk) / num_words) * chunk_length
                    ])
                    chunk_id += 1
            else:
                chunks.append([
                    chunk_id,
                    chunk_length,
                    text,
                    start_time,
                    end_time
                ])
                chunk_id += 1

        return chunks
    except Exception as e:
        # Return error message if an exception occurs
        return str(e)

def process_video(youtube_url):
    """
    Process the YouTube video: download, extract audio, transcribe, align, and chunk.

    Parameters:
    youtube_url (str): URL of the YouTube video.

    Returns:
    tuple: Status, transcript, and chunks.
    """
    video_path = download_video(youtube_url)
    if "Error" in video_path:
        return video_path, "", []

    audio_path = extract_audio(video_path)
    if "Error" in audio_path:
        return audio_path, "", []

    resampled_audio_path = resample_audio(audio_path)
    if "Error" in resampled_audio_path:
        return resampled_audio_path, "", []

    transcript = transcribe_audio(resampled_audio_path)
    if "Error" in transcript:
        return transcript, "", []

    alignment_path = align_transcript(resampled_audio_path, transcript)
    if "Error" in alignment_path:
        return alignment_path, "", []

    chunks = semantic_chunking(alignment_path)
    if isinstance(chunks, str) and "Error" in chunks:
        return chunks, "", []

    return "Success", transcript, chunks

# Provide the YouTube URL here to print semantic chunks
youtube_url = "https://www.youtube.com/watch?v=Sby1uJ_NFIY"
status, transcript, chunks = process_video(youtube_url)

if status == "Success":
    print("Transcription:")
    print(transcript)
    print("\nSemantic Chunks:")
    for chunk in chunks:
        print(f"Chunk ID: {chunk[0]}, Start Time: {chunk[3]}, End Time: {chunk[4]}, Text: {chunk[2]}")
else:
    print("Error:", status)
````

# **Utilizing Ground-Truth Transcripts:**

## Method to Improve Transcription Quality:

**Hypothesis:** Using a ground-truth transcript can significantly enhance the quality of the transcription by providing a reference that the model can use to correct errors.

**Approach:** Initial Transcription: Perform the initial transcription using the Whisper model.

**Alignment:** Use aeneas to align the initial transcript with the ground-truth transcript. This involves aligning the audio with the provided accurate transcript.

**Correction:** Use the alignment to correct the initial transcription. Segments of the initial transcription can be replaced or corrected based on the ground-truth alignment.

**Pseudo Code:**
```
INITIAL_TRANSCRIPTION_PROCESS:
    Input: Audio file
    Output: Initial transcript
    Steps:
        - Load Whisper model
        - Transcribe audio to generate initial transcript

SCRAPE_GROUND_TRUTH_TRANSCRIPT:
    Input: Ground-truth transcript URL
    Output: Ground-truth transcript text
    Steps:
        - Access website containing the ground-truth transcript
        - Extract the text of the transcript

ALIGN_TRANSCRIPTS:
    Input: Initial transcript, Ground-truth transcript, Audio file
    Output: Alignment data (sync map)
    Steps:
        - Use aeneas library to create an alignment task
        - Configure the task with initial transcript, ground-truth transcript, and audio file
        - Execute the alignment task to produce sync map

CORRECT_TRANSCRIPTION:
    Input: Initial transcript, Alignment data, Ground-truth transcript
    Output: Corrected transcript
    Steps:
        - For each segment in the alignment data:
            - Check if the initial transcript segment matches the ground-truth segment
            - If discrepancies found, correct the initial transcript segment using ground-truth segment

GENERATE_FINAL_OUTPUT:
    Input: Corrected transcript, Audio file
    Output: High-quality transcription with accurate segments
    Steps:
        - Ensure the corrected transcript aligns with the audio
        - Segment the corrected transcript into semantic chunks (if required)
```

**Consistency:** Ensures consistency in the transcription, particularly for critical or specialized terms that the model might misinterpret.

## Challenges:

**Alignment Accuracy:** Ensuring the alignment process is accurate and does not introduce new errors.

**Processing Time:** Additional steps for alignment and correction may increase processing time.

````python
# Web Scraping
import requests
from bs4 import BeautifulSoup
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import os
import asyncio
from pyppeteer import launch
import aiohttp

# Step 1: Fetch the main page and find the recordings database link
def fetch_recordings_database_url():
    """
    Fetches the recordings database URL from the Faith Comes By Hearing website.
    """
    main_url = "https://www.faithcomesbyhearing.com/"
    response = requests.get(main_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Locate the recordings database URL
    for link in soup.find_all('a', href=True):
        if 'recordings-database' in link['href']:
            return main_url.rstrip('/') + link['href']
    raise Exception("Recordings Database URL not found")

# Step 2: Use Selenium to find the Indian Revised Version link
def fetch_indian_revised_version_link(recordings_url):
    """
    Uses Selenium to find the Indian Revised Version link from the recordings database.
    """
    # Set up Chrome options for headless mode
    options = uc.ChromeOptions()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')

    # Add this line to set the binary location for Google Colab
    options.binary_location = "/usr/bin/google-chrome"

    # Initialize the Chrome driver
    driver = uc.Chrome(options=options)

    try:
        # Navigate to the recordings database page
        driver.get(recordings_url)
        print("Navigated to the recordings database page.")

        # Wait for the search input to load and type "Hindi"
        wait = WebDriverWait(driver, 20)
        search_input = wait.until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'input.default-input--search'))
        )
        search_input.send_keys('Hindi')
        print("Typed 'Hindi' in the search input.")

        # Wait for the table to update
        time.sleep(2)

        # Find the Indian Revised Version link
        links = driver.find_elements(By.CSS_SELECTOR, 'a.recordings-link')
        for link in links:
            if 'Indian Revised Version' in link.get_attribute('title'):
                print("Found the Indian Revised Version link.")
                return link.get_attribute('href')
        raise Exception("Indian Revised Version link not found.")
    finally:
        # Quit the driver
        driver.quit()
        print("Driver quit.")

# Step 3: Use Pyppeteer to scrape and process audio and text
async def get_html(page, url, retries=3, click_button=False):
    """
    Fetches the HTML content of a page, with optional retries and button clicks for dynamic content.
    """
    for attempt in range(retries):
        try:
            # Navigate to the specified URL
            await page.goto(url, {'waitUntil': 'networkidle2', 'timeout': 120000})

            if click_button:
                print("Ensuring correct page load...")
                await asyncio.sleep(2)  # Allow some time for scripts to load
                print("Clicking the visible chapter dropdown button...")
                await page.waitForSelector('div#chapter-dropdown-button', {'visible': True, 'timeout': 120000})
                await page.click('div#chapter-dropdown-button')
                print("Waiting for chapter text dropdown to be visible...")
                await page.waitForSelector('div.chapter-text-dropdown', {'visible': True, 'timeout': 120000})
                await asyncio.sleep(5)  # Adding a small delay to ensure the content is fully loaded

            # Return the page content
            html = await page.content()
            return html
        except Exception as e:
            print(f"Attempt {attempt + 1} failed with error: {e}")
            if attempt == retries - 1:
                raise
            await asyncio.sleep(2)

def extract_audio_and_text(soup):
    """
    Extracts the audio URL and chapter text from the BeautifulSoup object of the page.
    """
    # Find the audio player background and extract the audio URL
    audio_player_background = soup.find('div', class_='audio-player-background')
    if audio_player_background:
        video_tag = audio_player_background.find('video', class_='audio-player')
        if video_tag:
            audio_url = video_tag['src']
        else:
            print("No video tag found in audio player background.")
            audio_url = None
    else:
        print("No audio player background found.")
        audio_url = None

    # Extract the chapter text
    chapter_texts = []
    main_wrapper = soup.find('div', class_='chapter justify')
    if main_wrapper:
        for span in main_wrapper.find_all('span', class_='align-left'):
            chapter_texts.append(span.get_text(strip=True))
    chapter_text = ' '.join(chapter_texts)

    return audio_url, chapter_text

async def fetch_books(page):
    """
    Fetches the books and their chapters from the New Testament section.
    """
    # Navigate to the first book of the New Testament
    await page.goto('https://live.bible.is/bible/HINDPI/GEN/1', {'waitUntil': 'networkidle2', 'timeout': 500000})
    await page.waitForSelector('div#chapter-dropdown-button', {'visible': True, 'timeout': 500000})
    await page.click('div#chapter-dropdown-button')
    await page.waitForSelector('div.chapter-text-dropdown', {'visible': True, 'timeout': 500000})
    html = await page.content()
    soup = BeautifulSoup(html, 'html.parser')

    books = []
    new_testament = soup.find('h3', string=lambda text: 'New Testament' in text if text else False)
    if new_testament:
        print("Found New Testament section")
        book_divs = new_testament.find_all_next('div', class_='book-button')
        for book_div in book_divs:
            book_id = book_div['id']
            book_name = book_div.get_text(strip=True)
            chapter_links = [a['href'] for a in book_div.find_next('span', class_='chapter-container').find_all('a', class_='chapter-box')]
            books.append((book_name, book_id, chapter_links))
            print(f"Found book: {book_name} with ID: {book_id}, Chapters: {len(chapter_links)}")
    else:
        print("Could not find New Testament section")

    return books

async def download_file(session, url, dest):
    """
    Downloads a file from the given URL to the specified destination.
    """
    async with session.get(url) as response:
        with open(dest, 'wb') as f:
            f.write(await response.read())

async def process_chapter(page, book_name, book_id, chapter_url, session):
    """
    Processes each chapter by downloading its audio and text data.
    """
    # Create a folder for the book if it doesn't exist
    book_folder = os.path.join('New_Testament', book_name)
    os.makedirs(book_folder, exist_ok=True)

    chapter_number = chapter_url.split('/')[-1]
    full_chapter_url = f"https://live.bible.is{chapter_url}"

    retries = 10
    delay = 5

    while retries > 0:
        try:
            # Get the HTML content of the chapter page
            html = await get_html(page, full_chapter_url)
            soup = BeautifulSoup(html, 'html.parser')

            # Extract the audio URL and chapter text
            audio_url, chapter_text = extract_audio_and_text(soup)

            if audio_url or chapter_text:
                # Download the audio file if it exists
                if audio_url:
                    audio_dest = os.path.join(book_folder, f"Chapter_{chapter_number}.mp3")
                    await download_file(session, audio_url, audio_dest)
                    print(f"Audio saved to: {audio_dest}")

                # Save the chapter text to a file
                text_dest = os.path.join(book_folder, f"Chapter_{chapter_number}.txt")
                with open(text_dest, 'w', encoding='utf-8') as f:
                    f.write(chapter_text)
                print(f"Text saved to: {text_dest}")

                print(f"Book: {book_name}, Chapter: {chapter_number} - Successfully processed")
                break
            else:
                print(f"Retrying for chapter {chapter_number} of {book_name} due to no audio or text found")
                retries -= 1
                await asyncio.sleep(delay)
                delay *= 2
        except Exception as e:
            print(f"Failed to fetch chapter {chapter_number} of {book_name} on attempt: {5 - retries + 1}, Error: {e}")
            retries -= 1
            await asyncio.sleep(delay)
            delay *= 2

async def scrape_website(base_url):
    """
    Orchestrates the scraping of the website, including fetching books and processing chapters.
    """
    async with aiohttp.ClientSession() as session:
        browser = await launch(headless=True, args=['--no-sandbox', '--disable-setuid-sandbox'])
        page = await browser.newPage()

        books = await fetch_books(page)
        if not books:
            print("No books found. Exiting.")
            return

        for book_name, book_id, chapter_links in books:
            for chapter_url in chapter_links:
                await process_chapter(page, book_name, book_id, chapter_url, session)

            # Navigate back to New Testament section after each book
            print("Navigating back to New Testament section")
            await page.goto(base_url, {'waitUntil': 'networkidle2', 'timeout': 500000})
            await page.waitForSelector('div#chapter-dropdown-button', {'visible': True, 'timeout': 500000})
            await page.click('div#chapter-dropdown-button')
            await page.waitForSelector('div.chapter-text-dropdown', {'visible': True, 'timeout': 500000})

        await browser.close()

async def main():
    """
    Main function to execute the web scraping and processing.
    """
    # Fetch the recordings database URL
    recordings_url = fetch_recordings_database_url()
    print("Recordings Database URL:", recordings_url)

    # Fetch the Indian Revised Version link
    indian_revised_version_url = fetch_indian_revised_version_link(recordings_url)
    print(f"Indian Revised Version link: {indian_revised_version_url}")

    # Scrape the website using the Indian Revised Version link
    base_url = f"{indian_revised_version_url}/GEN/1"
    await scrape_website(base_url)

# Execute the main function
await main()
````

````python
import os
import pandas as pd
import librosa
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from langdetect import detect
from wordcloud import WordCloud
from sklearn.feature_extraction.text import CountVectorizer
from textblob import TextBlob
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import textstat
from aeneas.executetask import ExecuteTask
from aeneas.task import Task
import json
import tempfile


# Import the transliteration package
from indic_transliteration import sanscript
from indic_transliteration.sanscript import transliterate

# Function to transliterate Hindi terms to English
def transliterate_term(term):
    return transliterate(term, sanscript.DEVANAGARI, sanscript.ITRANS)


# Ensure matplotlib uses the correct font path
font_path = 'NotoSansDevanagari-Regular.ttf'

# Define the path to your dataset
base_path = '/content/New_Testament'

# Initialize lists to hold data
data = []
book_counter = 1

# Traverse through each book and chapter to load data
for book in os.listdir(base_path):
    book_path = os.path.join(base_path, book)
    if os.path.isdir(book_path):
        book_name = f'Book {book_counter}'
        book_counter += 1
        for chapter in os.listdir(book_path):
            if chapter.endswith('.mp3'):
                chapter_name = chapter[:-4]
                audio_path = os.path.join(book_path, chapter)
                text_path = os.path.join(book_path, chapter_name + '.txt')
                with open(text_path, 'r', encoding='utf-8') as file:
                    text = file.read()
                # Ensure the text is in Hindi
                if detect(text) != 'hi':
                    continue
                # Load audio file
                y, sr = librosa.load(audio_path, sr=None)
                duration = librosa.get_duration(y=y, sr=sr)
                data.append([book_name, chapter_name, audio_path, text, duration, sr])

# Create a DataFrame
df = pd.DataFrame(data, columns=['Book', 'Chapter', 'AudioPath', 'Text', 'Duration', 'SampleRate'])

# Function to extract audio features
def extract_audio_features(file_path):
    y, sr = librosa.load(file_path, sr=None)
    features = {
        'zero_crossing_rate': round(np.mean(librosa.feature.zero_crossing_rate(y=y)), 3),
        'spectral_centroid': round(np.mean(librosa.feature.spectral_centroid(y=y, sr=sr)), 3),
        'spectral_bandwidth': round(np.mean(librosa.feature.spectral_bandwidth(y=y, sr=sr)), 3),
        'rolloff': round(np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr)), 3),
        'mfcc': np.mean(librosa.feature.mfcc(y=y, sr=sr), axis=1).tolist(),
        'tempo': round(librosa.beat.tempo(y=y, sr=sr)[0], 3)
    }
    return features

# Extract features for all audio files
audio_features = df['AudioPath'].apply(extract_audio_features)
audio_features_df = pd.DataFrame(audio_features.tolist())
df = pd.concat([df, audio_features_df], axis=1)

# Audio Duration Analysis
plt.figure(figsize=(12, 6))
sns.histplot(df['Duration'], bins=50, kde=True)
plt.title('Distribution of Audio Durations')
plt.xlabel('Duration (seconds)')
plt.ylabel('Frequency')
plt.show()

# Sample Rate Analysis
plt.figure(figsize=(12, 6))
sns.countplot(x='SampleRate', data=df)
plt.title('Sample Rate Distribution')
plt.xlabel('Sample Rate')
plt.ylabel('Count')
plt.show()

# Visualize some audio features
plt.figure(figsize=(12, 6))
sns.boxplot(x='Book', y='spectral_centroid', data=df)
plt.xticks(rotation=90)
plt.title('Spectral Centroid by Book')
plt.show()

# Text Length Analysis
df['TextLength'] = df['Text'].apply(len)
plt.figure(figsize=(12, 6))
sns.histplot(df['TextLength'], bins=50, kde=True)
plt.title('Distribution of Text Lengths')
plt.xlabel('Text Length (characters)')
plt.ylabel('Frequency')
plt.show()

# Word Cloud for a sample book
sample_book_text = ' '.join(df[df['Book'] == df['Book'].iloc[0]]['Text'].tolist())
wordcloud = WordCloud(width=800, height=400, background_color='white', font_path=font_path).generate(sample_book_text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for a Sample Book')
plt.show()

# Term Frequency Analysis
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['Text'])
terms = vectorizer.get_feature_names_out()
term_freq = np.asarray(X.sum(axis=0)).flatten()
term_freq_df = pd.DataFrame({'term': terms, 'frequency': term_freq})
term_freq_df = term_freq_df.sort_values(by='frequency', ascending=False)

# Top 20 Terms
top_terms = term_freq_df.head(20)
top_terms['term'] = top_terms['term'].apply(transliterate_term)
plt.figure(figsize=(12, 6))
sns.barplot(x='frequency', y='term', data=top_terms)
plt.title('Top 20 Terms in Text Data')
plt.xlabel('Frequency')
plt.ylabel('Term')
plt.show()

# Sentiment Analysis
df['Sentiment'] = df['Text'].apply(lambda x: TextBlob(x).sentiment.polarity)
plt.figure(figsize=(12, 6))
sns.histplot(df['Sentiment'], bins=50, kde=True)
plt.title('Sentiment Analysis of Text')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.show()

# Readability Analysis
df['Readability'] = df['Text'].apply(textstat.flesch_reading_ease)
plt.figure(figsize=(12, 6))
sns.histplot(df['Readability'], bins=50, kde=True)
plt.title('Readability Scores of Text')
plt.xlabel('Readability Score')
plt.ylabel('Frequency')
plt.show()

# Additional KPIs
kpis = {
    'Average Audio Duration': round(df['Duration'].mean(), 3),
    'Median Audio Duration': round(df['Duration'].median(), 3),
    'Standard Deviation of Audio Duration': round(df['Duration'].std(), 3),
    'Longest Audio': round(df['Duration'].max(), 3),
    'Shortest Audio': round(df['Duration'].min(), 3),
    'Sample Rate Consistency': df['SampleRate'].nunique() == 1,
    'Average Text Length': round(df['TextLength'].mean(), 3),
    'Text-Audio Correlation': round(df[['TextLength', 'Duration']].corr().iloc[0, 1], 3),
    'Average Spectral Centroid': round(df['spectral_centroid'].mean(), 3),
    'Average Spectral Bandwidth': round(df['spectral_bandwidth'].mean(), 3),
    'Average MFCC': [round(np.mean(mfcc), 3) for mfcc in zip(*df['mfcc'])],
    'Tempo': round(df['tempo'].mean(), 3),
    'Average Sentiment': round(df['Sentiment'].mean(), 3),
    'Average Readability': round(df['Readability'].mean(), 3),
}

print("Key Performance Indicators:")
for key, value in kpis.items():
    print(f"{key}: {value}")

# Clustering Analysis
audio_features_clustering = df[['zero_crossing_rate', 'spectral_centroid', 'spectral_bandwidth', 'rolloff']]
scaler = StandardScaler()
audio_features_scaled = scaler.fit_transform(audio_features_clustering)
pca = PCA(n_components=2)
audio_features_pca = pca.fit_transform(audio_features_scaled)
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(audio_features_pca)
df['Cluster'] = clusters

plt.figure(figsize=(12, 6))
sns.scatterplot(x=audio_features_pca[:, 0], y=audio_features_pca[:, 1], hue=df['Cluster'], palette='Set1')
plt.title('Clustering of Audio Features')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# Phoneme alignment using aeneas
def phoneme_alignment(text, audio_path):
    with tempfile.NamedTemporaryFile(delete=False, mode='w', encoding='utf-8') as temp_text_file:
        temp_text_file.write(text)
        temp_text_file_path = temp_text_file.name

    config_string = "task_language=hin|os_task_file_format=json|is_text_type=plain"
    task = Task(config_string=config_string)
    task.audio_file_path_absolute = audio_path
    task.text_file_path_absolute = temp_text_file_path

    task.sync_map_file_path_absolute = "output_syncmap.json"

    ExecuteTask(task).execute()
    task.output_sync_map_file()

    with open("output_syncmap.json", "r", encoding="utf-8") as f:
        sync_map = json.load(f)

    alignment_score = np.mean([float(fragment["end"]) - float(fragment["begin"]) for fragment in sync_map["fragments"]])

    os.remove(temp_text_file_path)

    return alignment_score

df['PhonemeAlignment'] = df.apply(lambda row: phoneme_alignment(row['Text'], row['AudioPath']), axis=1)
plt.figure(figsize=(12, 6))
sns.histplot(df['PhonemeAlignment'], bins=50, kde=True)
plt.title('Phoneme Alignment Scores')
plt.xlabel('Alignment Score')
plt.ylabel('Frequency')
plt.show()

# Temporal Analysis
# Adjusting the chapter labels for readability
plt.figure(figsize=(16, 10))
sns.lineplot(data=df, x='Chapter', y='TextLength', hue='Book', marker='o')
plt.title('Text Length Over Chapters')
plt.xlabel('Chapter')
plt.ylabel('Text Length')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(16, 10))
sns.lineplot(data=df, x='Chapter', y='Sentiment', hue='Book', marker='o')
plt.title('Sentiment Over Chapters')
plt.xlabel('Chapter')
plt.ylabel('Sentiment')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90)
plt.show()

# Save DataFrame for further analysis
df.to_csv('new_testament_analysis.csv', index=False)
````

        chapter_text = ' '.join(chapter_texts)

    async def fetch_books(page):

            for book_div in book_divs:

                book_name = book_div.get_text(strip=True)
                chapter_links = [a['href'] for a in book_div.find_next('span', class_='chapter-container').find_all('a', class_='chapter-box')]
                books.append((book_name, book_id, chapter_links))
        return books

    async def download_file(session, url, dest):

                f.write(await response.read())

    async def process_chapter(page, book_name, book_id, chapter_url, session):
        book_folder = os.path.join('New_Testament', book_name)
        os.makedirs(book_folder, exist_ok=True)
        chapter_number = chapter_url.split('/')[-1]
        full_chapter_url = f"https://live.bible.is{chapter_url}"
        retries = 10
        delay = 5

        while retries > 0:

                    text_dest = os.path.join(book_folder, f"Chapter_{chapter_number}.txt")
                    with open(text_dest, 'w', encoding='utf-8') as f:
                        f.write(chapter_text)
                    break
                else:
                    retries -= 1
                    await asyncio.sleep(delay)
                    delay *= 2
            except Exception as e:
                retries -= 1
                await asyncio.sleep(delay)
                delay *= 2
